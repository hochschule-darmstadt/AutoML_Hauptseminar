```python
# import libraries
import pandas as pd
import re
import autosklearn.classification
import autosklearn.regression
import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics
from sklearn.metrics import mean_squared_error, f1_score
```

```python
# load datasets
college_train = pd.read_csv('/opt/nb/Desktop/DS3/AutoML/AutoML_Hauptseminar/Hauptseminar-AutoKeras/college_train.csv', quotechar='\'', skipinitialspace=True, na_values="?")
college_test = pd.read_csv('/opt/nb/Desktop/DS3/AutoML/AutoML_Hauptseminar/Hauptseminar-AutoKeras/college_test.csv', quotechar='\'', skipinitialspace=True, na_values="?")
phishing_train = pd.read_csv('/opt/nb/Desktop/DS3/AutoML/AutoML_Hauptseminar/Hauptseminar-AutoKeras/phishing_train.csv', quotechar='\'', skipinitialspace=True, na_values="?")
phishing_test = pd.read_csv('/opt/nb/Desktop/DS3/AutoML/AutoML_Hauptseminar/Hauptseminar-AutoKeras/phishing_test.csv', quotechar='\'', skipinitialspace=True, na_values="?")
```

Classification dataset

```python
X_train = phishing_train.iloc[:,:-1] 
y_train = phishing_train.iloc[:,-1] # target is Result
X_test = phishing_test.iloc[:,:-1]
y_test = phishing_test.iloc[:,-1]
```

```python
# 1st predictions and print f1-score
cls = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=600)
cls.fit(X_train, y_train)
y_hat = cls.predict(X_test)
test_f1 = f1_score(y_test, y_hat)
print(f"Test F1-Score {test_f1}")

# 2nd predictions and print f1-score
cls = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=600, seed=142)
cls.fit(X_train, y_train)
y_hat = cls.predict(X_test)
test_f1 = f1_score(y_test, y_hat)
print(f"Test F1-Score {test_f1}")

# 3rd make predictions and print f1-score
cls = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=600, seed=143)
cls.fit(X_train, y_train)
y_hat = cls.predict(X_test)
test_f1 = f1_score(y_test, y_hat)
print(f"Test F1-Score {test_f1}")
```

```python
# remove quotes from column names
college_train.columns = college_train.columns.str.replace('"', '')
college_test.columns = college_test.columns.str.replace('"', '')
```

```python
# get number of nuniques of categorical variables
for x in college_train[["school_name", "city", "zip", "school_webpage", "state", "predominant_degree", "highest_degree", "ownership", "region", "gender", "carnegie_basic_classification", "carnegie_undergraduate", "carnegie_size", "religious_affiliation"]]:
    print(f"nunique {x}:  college_train: {len(college_train[x].unique())} | college_test: {len(college_test[x].unique())}")
```

```python
# combined dataset
combined = pd.concat([college_train, college_test])
```

```python
# define next steps for categorical variables:
to_change= ["state", "predominant_degree","highest_degree","ownership","region","gender","carnegie_basic_classification","carnegie_undergraduate","carnegie_size","religious_affiliation"]
to_drop= ["school_name", "city", "zip", "school_webpage"]
```

```python
# remove columns with more than 100 unqiue values on categorical variables for one-hot-encoding
combined.drop(to_drop, axis=1, inplace=True)
```

```python
print(f"nrows college_train: {len(college_train)}")
print(f"nrows college_test: {len(college_test)}")
```

```python
# use one hot encoding for categorical variables
combined = pd.get_dummies(combined, prefix=to_change)
```

```python
# resplit to train/test sets
college_train = combined.iloc[:4944,:]
college_test = combined.iloc[4944:,:]
```

```python
X_train = college_train.loc[:, college_train.columns != "percent_pell_grant"]
y_train = college_train.loc[:, college_train.columns == "percent_pell_grant"]
X_test = college_test.loc[:, college_test.columns != "percent_pell_grant"]
y_test = college_test.loc[:, college_test.columns == "percent_pell_grant"]
```

```python
# 1st automl on regression dataset
cls = autosklearn.regression.AutoSklearnRegressor(time_left_for_this_task=600, seed=141)
cls.fit(X_train, y_train)
y_hat = cls.predict(X_test)
rmse=mean_squared_error(y_test, y_hat)**0.5
print('RMSE: ' + str(rmse))

# 2nd automl on regression dataset
cls = autosklearn.regression.AutoSklearnRegressor(time_left_for_this_task=600, seed=142)
cls.fit(X_train, y_train)
y_hat = cls.predict(X_test)
rmse=mean_squared_error(y_test, y_hat)**0.5
print('RMSE: ' + str(rmse))

# 3rd automl on regression dataset
cls = autosklearn.regression.AutoSklearnRegressor(time_left_for_this_task=600, seed=143)
cls.fit(X_train, y_train)
y_hat = cls.predict(X_test)
rmse=mean_squared_error(y_test, y_hat)**0.5
print('RMSE: ' + str(rmse))
```
